```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot scatter
def plot_samples(samples, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(samples[:, 0], samples[:, 1], alpha=0.6)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(title)
    plt.xlabel("$x_1$")
    plt.ylabel("$x_2$")
    plt.grid(alpha=0.3)
    plt.show()

# 1. Mean [0, 0] and Identity Covariance Matrix
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # Identity matrix
samples = np.random.multivariate_normal(mean, cov, 100)
plot_samples(samples, "Mean: [0, 0], Covariance: Identity Matrix")

# 2. Mean [1, 1]
mean = [1, 1]
samples = np.random.multivariate_normal(mean, cov, 100)
plot_samples(samples, "Mean: [1, 1], Covariance: Identity Matrix")

# 3. Double the Variance
mean = [0, 0]
cov = [[2, 0], [0, 2]]  # Double variance
samples = np.random.multivariate_normal(mean, cov, 100)
plot_samples(samples, "Mean: [0, 0], Covariance: Double Variance")

# 4. Covariance Matrix with Correlation
cov = [[1, 0.5], [0.5, 1]]
samples = np.random.multivariate_normal(mean, cov, 100)
plot_samples(samples, "Mean: [0, 0], Covariance: Positive Correlation")

# 5. Negative Correlation
cov = [[1, -0.5], [-0.5, 1]]
samples = np.random.multivariate_normal(mean, cov, 100)
plot_samples(samples, "Mean: [0, 0], Covariance: Negative Correlation")
```
